```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función logarítmica
def f(x):
    return np.log(x)

# Crear un rango de valores para x, evitando el 0 ya que log(0) no está definido
x = np.linspace(0.1, 10, 400)

# Calcular los valores de y usando la función
y = f(x)

# Crear la gráfica
plt.plot(x, y, label='f(x) = log(x)')

# Añadir título y etiquetas
plt.title('Gráfica de la función f(x) = log(x)')
plt.xlabel('x')
plt.ylabel('f(x)')

# Añadir una leyenda
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones trigonométricas
def seno(x):
    return np.sin(x)

def coseno(x):
    return np.cos(x)

def tangente(x):
    return np.tan(x)

# Crear un rango de valores para x
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# Calcular los valores de y usando las funciones
y_seno = seno(x)
y_coseno = coseno(x)
y_tangente = tangente(x)

# Crear la gráfica
plt.figure(figsize=(10, 6))

# Graficar la función seno
plt.plot(x, y_seno, label='seno(x)', color='blue')

# Graficar la función coseno
plt.plot(x, y_coseno, label='coseno(x)', color='red')

# Graficar la función tangente
plt.plot(x, y_tangente, label='tangente(x)', color='green')

# Añadir título y etiquetas
plt.title('Gráficas de las funciones trigonométricas')
plt.xlabel('x')
plt.ylabel('f(x)')

# Añadir una leyenda
plt.legend()

# Añadir una cuadrícula
plt.grid(True)

# Mostrar la gráfica
plt.ylim(-10, 10)  # Limitar el eje y para evitar valores extremos de la tangente
plt.show()
```

```python
import sympy as sp

# Define la variable simbólica
x = sp.symbols('x')

# Define la función logarítmica
f = sp.log(x)

# Calcula el límite cuando x tiende a 0 por la derecha
limite_derecha = sp.limit(f, x, 0, dir='+')

# Calcula el límite cuando x tiende a infinito
limite_infinito = sp.limit(f, x, sp.oo)

print(f"Límite de log(x) cuando x tiende a 0 por la derecha: {limite_derecha}")
print(f"Límite de log(x) cuando x tiende a infinito: {limite_infinito}")
```
